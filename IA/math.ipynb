### Libreria math

```python
import math

a = math.factorial(3)
b = math.sqrt(4)
c = math.ceil(4.12)
d = math.floor(4.12)

print(f"Factorial de 3 = {a}")
print(f"Raiz de 4 = {b}")
print(f"Ceil de 4.12 = {c}")
print(f"floor de 4.12 = {d}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import pi

x = np.linspace(0, 4*pi, 40)
y = np.sin(x)
z = np.cos(x)

plt.scatter(x, y, marker='*', color='red')
plt.plot(x, z, marker="+", color='blue')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
