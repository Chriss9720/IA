```python
import numpy as np
import matplotlib.pyplot as plt
Y=pd.read_csv("./data/Tarea4Ejemplo2Datos2.csv")
Y = Y.to_numpy()
datos = pd.read_csv("./data/Tarea4Ejemplo2Datos1.csv")
datos = datos.to_numpy()
datosx = pd.read_csv("./data/Tarea4Ejemplo2Datos3.csv")
datosx = datosx.to_numpy()

x=datosx
A = datos

print('A= ',A,sep='\n',end='\n\n')
print('Y= ',Y,sep='\n',end='\n\n')


X= np.linalg.inv(A.T.dot(A)).dot(A.T).dot(Y)

y=((2000)+(1000*x))
err=(y-Y)
errCuad=(err*err)

mse = (np.square(err - errCuad)).mean()
print("Valor de Y Prima \n")
print(y)
print("\nValor del error normal ")
print(err)
print("\nValor del error cuadratico")
print(errCuad)
print('X= ',X,sep='\n',end='\n\n')
print("\nError CUADRATICO MEDIO")
print(mse)

fig1 = plt.figure()
plt.plot(Y,A[:,1], 'r-')
plt.ylabel('salida sensor')
plt.xlabel('nivel real')
plt.title('Ecuacion Lineal')
Salida= A.dot(X)
plt.plot(Salida,A[:,1], 'b-')
```
